```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pickle
```

```python
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
#Preprocess

data= data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data.head()
```

```python
#Transform Categorical features
label_encoder= LabelEncoder()
data['Gender']=label_encoder.fit_transform(data['Gender'])
data
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder= OneHotEncoder()
geo=one_hot_encoder.fit_transform(data[['Geography']])

geo
```

```python
one_hot_encoder.get_feature_names_out(['Geography'])
one_hot_geo_df=pd.DataFrame(geo.toarray(),columns=one_hot_encoder.get_feature_names_out(['Geography'])
)
```

```python
data= pd.concat([data.drop('Geography',axis=1),one_hot_geo_df],axis=1)
data
```

```python
data
```

```python
with open('label_encoder.pkl','wb') as file:
    pickle.dump(label_encoder,file)

with open('one_hot_encoderv2.pkl','wb') as file:
    pickle.dump(one_hot_encoder,file)
```

```python
with open('one_hot_encoderv2.pkl','wb') as file:
    pickle.dump(one_hot_encoder,file)
```

```python
data.head()
X= data.drop('Exited',axis=1)
y= data['Exited']

x_train, x_test, y_train, y_test= train_test_split(X,y,random_state=42,test_size=0.2)
scaler= StandardScaler()
x_train= scaler.fit_transform(x_train)
x_test= scaler.transform(x_test)
```

```python
x_train
```

```python
with open('scaler.pkl', 'wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
x_train.shape[1]
X.head()
```

```python
model = Sequential([Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
Dense(32,activation='relu'),
Dense(1,activation='sigmoid')
])
model.summary()
```

```python
import tensorflow as tf
opt=  tf.keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
logs_dir="ANNlogs/log/"
tensorflow_callback= TensorBoard(log_dir=logs_dir,histogram_freq=1)
```

```python
early_stopping_callback= EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history= model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100,callbacks=[tensorflow_callback,early_stopping_callback])
```

```python
model.save('model.h5')
```
